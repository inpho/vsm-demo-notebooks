# Refactored Viewer Tutorial  (9/28/2013)

```python
# importing model and corpus
from vsm.corpus import Corpus
from vsm.model.ldagibbs import LDAGibbs as LDA
from vsm.viewer.ldagibbsviewer import LDAGibbsViewer as LDAViewer

c = Corpus.load('/var/inphosemantics/data/fresh/sep/vsm-corpora/sep-freq1-nltk-compressed.npz')

# 100-topic model. Alternative choices are: "sep-LDA-K{40, 60, 80, 100, 120, 140, 160}-2000.npz" 
m = LDA.load('/var/inphosemantics/data/fresh/sep/vsm-matrices/sep-LDA-K100-2000.npz')
v = LDAViewer(c,m)
```

## change 1: topic analysis is now based on a `Topics` class

```python
# first you create a Topics class
t = v.topics()   # <- this creates a Topics class including all topics in the viewer
```

```python
# to view the topics, use .view() method
t.view()
```

```python
# all subsequent analysis are also done by Topics methods.
t.dismat() # calculates and stores distance between topics
t.isomap() # creates an isomap based on the distance 
t.KMeans() # cluster topics based on the above
t.plot()   # plot topics based on the above
```

```python
# all cluster infomation is in `.cls` To see the contents of the 4th cluster,  
t.view(t.cls[0])
```

## Changes 2: every sim_ function now creates classes on which you can do various analyses

```python
# first let's create top 10 topics relevant to `kant`
t_kant = v.sim_word_top('kant', print_len=10) 
t_kant
```

```python
# you can also view the contents
t_kant.view()
```

```python
t_kant.dismat()
t_kant.isomap()
t_kant.plot()
```

```python
# Next let's create top 100 documents relevant to these topics
d_kant = v.sim_top_doc(t_kant.labels, print_len=100)
d_kant
```

```python
# Visualization is easy
d_kant.dismat()
d_kant.isomap()
d_kant.KMeans()
d_kant.plot()
```
