```python
# importing necessary libraries #

from vsm.extensions.htrc import *
from vsm.corpus.util.corpusbuilders import *
from vsm.corpus import Corpus
```

```python
# Corpus cleanup for HTRC #
# Cleans up page headers, line breaks, and hyphens for all plain pages in the book directory.
# Creates a log file for debugging purposes. 
# ignore is a list of file extensions to ignore in the directory.

plain_dir = '/var/inphosemantics/data/extensions/'
book = 'uc2.ark+=13960=t5w66bs1h'

proc_htrc_book(book, plain_dir, ignore=['.json', '.log'])
```

```python
# Build a Corpus object #
# chunk_name is the name of tokenization corresponding to individual files.
# stop_freq eliminates words that appear <= 3 times.
full_path = plain_dir + book

c = dir_corpus(full_path, chunk_name='page', nltk_stop=True, stop_freq=3, ignore=['.json', '.pickle', '.xml'])
```

```python
# save the Corpus object as a .npz file.
c.save(plain_dir + 'uc2.ark+=13960=t5w66bs1h-nltk-freq3.npz')
```

```python
# Saved Corpus object can be loaded again.

c1 = Corpus.load(plain_dir + 'uc2.ark+=13960=t5w66bs1h-nltk-freq3.npz')
```
