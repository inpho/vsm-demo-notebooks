# LDA Tutorial for new manifold learnings 

```python
# importing model and corpus
from vsm.corpus import Corpus
from vsm.model.ldagibbs import LDAGibbs as LDA
from vsm.viewer.ldagibbsviewer import LDAGibbsViewer as LDAViewer

c = Corpus.load('/var/inphosemantics/data/fresh/sep/vsm-corpora/sep-freq1-nltk-compressed.npz')
m = LDA.load('/var/inphosemantics/data/fresh/sep/vsm-matrices/sep-LDA-K100-2000.npz')
v = LDAViewer(c,m)
```

# Create a dissimilarity matrix

```python
mat = v.dismat_topics()  # of topics. 
# mat = v.dismat_docs()  # of docs. !!Takes time!!
# mat = v.dismat_words()  # of words. !!Takes time!! 
```

```python
# to view the matrix
mat.dismat
```

# Low-dimensionality embedding

This basically embeds objects (topics, docs, or words) into a fewer dimensional space. The new coordinates in the reduced dimsntion are stored into `mat.pos`.

```python
mat.isomap(n_components=2, n_neighbors=3) # reduction to 2d with isomap 
#mat.mds(n_components=2)    # with mds
```

```python
mat.pos
```

# Manifold learning

```python
mat.KMeans(n_clusters=10)      # KMeans. Needs to run mds or isomap first.

#mat.SpectralClustering(n_clusters=10) # spectral clustering
#mat.AffinityPropagation()      # Affinity propagation 
```

```python
# you can review the topics in each cluster
v.topics(k_indices=mat.cls[0])
```
