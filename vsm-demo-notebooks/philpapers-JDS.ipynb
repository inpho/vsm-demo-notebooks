# KL-divergence and JS-divergence demo

```python
from vsm.corpus import Corpus
from vsm.model.ldacgsmulti import LdaCgsMulti as Lda
from vsm.viewer.ldagibbsviewer import LDAGibbsViewer as LdaViewer
```

```python
c = Corpus.load('/var/inphosemantics/data/20130101/philpapers/vsm-data/'
                'philpapers-nltk-freq1.corpus.npz')
```

```python
m = Lda.load('/var/inphosemantics/data/20130101/philpapers/vsm-data/'
             'philpapers-nltk-freq1-ldacgsmulti-K140-1500-1.model.npz')
v = LdaViewer(c, m)
```

# sim functions

```python
v.sim_doc_doc(doc_or_docs='OCOPCF.json')  #"O'Connor, Timothy (2005). ""Pastoral counsel for the anxious naturalist: Daniel Dennett's freedom evolves"""
```

```python
v.sim_top_top(topic_or_topics=[1])
```

```python
v.topics(k_indices=[1,111,7,43,133,80,81,125,78,101])
```

# Simmat function

```python
# using Jensen-Shannon Divergence --- slower than cosine
time v.simmat_topics(measure='JSD')
```

```python
time v.simmat_topics(measure='cosine')
```

# Isomap_topics

```python
# Using Jensen-Shannon Divergence
cls_JSD = v.cluster_topics(measure='JSD')
v.isomap_topics(k_indices=cls_JSD, measure='JSD')
```

```python
# Using cosine
cls_cos = v.cluster_topics(measure='cosine')
v.isomap_topics(k_indices=cls_cos, measure='cosine')
```
