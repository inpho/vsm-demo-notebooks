# Topic Model Stability

Using the AP89 corpus, I preformed an analysis over the bag of words to see how multiple runs for the same number of k varied. I did this by looking at the log likelihoods of the models throughout the training process, and then did topic alignments between pairs of models. These topic models were carried out using several different methods of 1-1 mapping, using several heuristics from chapter 16 on text clustering.

## Topic Alignments

Finally, I will have a topic alignment consisting of the most similar element by JSD. I will implement several conflict resolution strategies for discussion. Topic alignments will be evaluated by taking the sum of the JSD between each alignment pair.

```python
def alignment_fitness(topic_pairs, dist=None, dist_fn=JS_dist):
    """
    Takes a list of topic pair tuples and returns the sum of the JS_dist between them
    """
    if dist is None:
        dist = model_dist(v1, v2, dist_fn)
    
    return sum([dist[t[0]][t[1]] for t in topic_pairs])

def plot_alignment(v1, v2, dist, alignment=None, fn_name=None):
    # Calculate distance between all topics
    Xs = cm.jet_r(dist)
    
    if alignment is None:
        alpha = 1.0
    else:
        alpha = np.zeros(dist.shape)
        alpha[zip(*alignment)] = 1
    
    Xs[:,:,3] = alpha
    
    # plot distance heatmap
    figure(figsize=(12,10))
    imshow(Xs, interpolation='nearest', cmap='jet_r', vmin=0, vmax=1.0)
    colorbar()
    #imshow(alpha, interpolation=None, cmap=get_cmap('binary'), vmin=0, vmax=1.0, alpha=0.5)
    xticks(np.arange(dist.shape[1]))
    yticks(np.arange(dist.shape[0]))
    
    if fn_name is not None:
        title("Topic Alignment %s() using Jensen-Shannon Distance" % fn_name)
    else:
        title("Topic Alignment")
    xlabel("Model 1")
    ylabel("Model 2")
    show()
```

```python
def basic_alignment(v1, v2, dist=None, dist_fn=JS_dist, debug=False):
    if dist is None:
        dist = model_dist(v1, v2, dist_fn)
        
    alignment = []
    for i, topic in enumerate(dist):
        # topic = a[i]
        #s = topic[topic.argsort()]
        #sim = topic.argsort()[s < 0.05]
        closest = topic.argsort()[0]
        alignment.append((i, closest))
        if debug:
            print i, closest, topic[closest]
    
    return alignment
```

```python
def naive_no_overlap(v1, v2, dist=None, dist_fn=JS_dist, debug=False):
    """
    First naive overlap detector just goes to next one if the topic has already been assigned
    """
    if dist is None:
        dist = model_dist(v1, v2, dist_fn)
        
    alignment = []
    aligned = []
    for i, topic in enumerate(dist):
        # topic = a[i]
        #s = topic[topic.argsort()]
        #sim = topic.argsort()[s < 0.05]
        topic_idx = 0
        closest = topic.argsort()[topic_idx]
        if debug:
            print i, closest, topic[closest]
        
        while closest in aligned:
            topic_idx += 1
            closest = topic.argsort()[topic_idx]
            if debug:
                print i, closest, topic[closest]
        
        
        aligned.append(closest)
        alignment.append((i, closest))
    
    return alignment
```

```python
from collections import defaultdict
def align_closest_old(v1,v2, dist=None, dist_fn=JS_dist, debug=False):
    if dist is None:
        dist = model_dist(v1, v2, dist_fn)

    # first populate basic alignment
    alignment = basic_alignment(v1,v2,dist, debug=debug)

    closest_idx = 0
    # now gather conflicts
    while True:
        rev_alignment = defaultdict(list)
        for pair in alignment:
            rev_alignment[pair[1]].append(pair[0])
        conflicts = filter(lambda x: len(x[1]) > 1, rev_alignment.items())
    
        if not conflicts:
            break
    
        closest_idx += 1
        # now assign permanently to the minimum and reassign the others to next closest
        for m2_idx, m1_idxs in conflicts:
            # sort by distance to m2
            m1_idxs = sorted(m1_idxs, key=lambda m1_idx: dist[m1_idx][m2_idx])
        
            # maintain alignment of closest element
            alignment[m1_idxs[0]] = (m1_idxs[0], m2_idx)
        
            # realign all others
            for other in m1_idxs[1:]:
                topic = dist[other]
                next_closest = topic.argsort()[closest_idx]
                alignment[other] = (other, next_closest)
                if debug:
                    print other, next_closest, topic[next_closest]
        
    return alignment

from collections import defaultdict
def align_closest(v1,v2, dist=None, dist_fn=JS_dist, debug=False):
    if dist is None:
        dist = model_dist(v1, v2, dist_fn)

    # first populate basic alignment
    alignment = basic_alignment(v1,v2,dist, debug=debug)

    next_closest_idx = defaultdict(lambda: 1)
    # now gather conflicts
    while True:
        rev_alignment = defaultdict(list)
        for pair in alignment:
            rev_alignment[pair[1]].append(pair[0])
        conflicts = filter(lambda x: len(x[1]) > 1, rev_alignment.items())
    
        if not conflicts:
            break
    
        # now assign permanently to the minimum and reassign the others to next closest
        for m2_idx, m1_idxs in conflicts:
            # sort by distance to m2
            m1_idxs = sorted(m1_idxs, key=lambda m1_idx: dist[m1_idx][m2_idx])
        
            # maintain alignment of closest element
            alignment[m1_idxs[0]] = (m1_idxs[0], m2_idx)
        
            # realign all others
            for other in m1_idxs[1:]:
                topic = dist[other]
                next_closest = topic.argsort()[next_closest_idx[other]]
                alignment[other] = (other, next_closest)
                next_closest_idx[other] += 1
                if debug:
                    print other, next_closest, topic[next_closest]
        
    return alignment
```

## Comparing alignments

```python
alignment_fns = [basic_alignment, naive_no_overlap, 
                 align_closest, align_second, align_min_neighborhood, 
                 align_second_diff, align_min_neighborhood_diff]

def compare_alignments(v1,v2):
    dist = model_dist(v1, v2, JS_dist)
    alignments = [fn(v1,v2, dist) for fn in alignment_fns]
    fitness = [alignment_fitness(alignment, dist) for alignment in alignments]
    pairings = [(fn.__name__, alignment, fitness) for fn, alignment, fitness in zip(alignment_fns, alignments, fitness)]
    
    # find minimum alignment, trimming basic_alignment
    for alignment in sorted(pairings, key=lambda x: x[2]):
        print alignment[0], alignment[2]
    min_alignment = min(pairings[1:], key=lambda x: x[2])
    plot_alignment(v1,v2, dist, alignment=min_alignment[1], fn_name=min_alignment[0])
```
