# Accessing doc-topic probabilities

There are multiple ways to access the doc-topic matrix using the Topic Explorer:
  - `v.doc_topics()`
  - `v.doc_topic_matrix()`

Each of these has a slightly different functionality that may be useful.

To use this demonstration notebook, upload it to your notebooks folder generated by the `vsm notebook` command then hit `Cell -> Run All` in the menus above.

```python
from corpus import *
k = topic_range[0]     # OPTIONAL: replace `topic_range[0]` with a specific number of topics (e.g. `k = 20`)
v = lda_v[k]
```

## `v.doc_topics()`

`v.doc_topics(doc_or_docs)` prints a notebook-friendly version of the topics. For outputting to a csv file, we recommend the `v.doc_topic_matrix()` function.

The single parameter `doc_or_docs` specifies which elements to return. The most common usage is to return all documents. the set of all document IDs is stored in the `v.labels` parameter.

```python
v.doc_topics(v.labels)
```

## `v.doc_topic_matrix()`

`v.doc_topic_matrix(doc_or_docs)` returns the doc_topic matrix without column or row headers. This can be easily used to output a csv file.

Again, `doc_or_docs` specifies which elements to return. The most common usage is to return all documents. the set of all document IDs is stored in the `v.labels` parameter.

```python
v.doc_topic_matrix(v.labels)
```

```python
# Writing a CSV file of doc_topic matrix with labels
with open('theta.{}.csv'.format(k), 'w') as csvfile:
    # print column headers
    header = 'ID,'
    header += ','.join(['T{}'.format(i) for i in range(k)]) + '\n'
    csvfile.write(header)
    
    for label, doc_topics in zip(v.labels, v.doc_topic_matrix(v.labels)):
        row = label + ','
        row += ','.join(['{}'.format(topic) for topic in doc_topics]) + '\n'
        csvfile.write(row)
```
